```python
import numpy_financial as npf
import matplotlib.pyplot as plt
import numpy as np
# Data:
price_t0 = 120000
percentage_per_year = 0.05
nominal_interest_rate = 0.12
period = 5
```

# 1. Price of the apartment after 5 years

```python
price_after_five_years = npf.fv(percentage_per_year/12, period*12, 0, -price_t0)
price_after_five_years
```

# 2. How much to save per month in order to collect money within 5 years

```python
monthly_payment = npf.pmt(nominal_interest_rate/4, period*12, price_after_five_years)
monthly_payment
```

# 3. Plot- apartment price/month; value of the investment

```python
# value of the investment
freq = 12
period = 5
nper = period * freq
periods = np.arange(1,nper+1,dtype=int)
apartment_price = -npf.ppmt(percentage_per_year/12, periods, period*12, price_after_five_years)
plt.plot(apartment_price.cumsum(),label='Apartment price')
value_of_the_investment = -npf.ppmt(nominal_interest_rate/4,periods,period*12, price_after_five_years)
plt.plot(value_of_the_investment.cumsum(),label='Value of the investment')
plt.legend()
plt.xlabel('time [month]')
plt.ylabel('Value [PLN]')
```
