```python
import numpy_financial as npf
import matplotlib.pyplot as plt
import numpy as np
# Data:
price_t0 = 120000
percentage_per_year = 0.05
nominal_interest_rate = 0.12
period = 5
```

# 1. Price of the apartment after 5 years

```python
price_after_five_years = npf.fv(percentage_per_year, period, 0, price_t0)
price_after_five_years
```

# 2. How much to save per month in order to collect money within 5 years

```python
monthly_payment = npf.pmt(nominal_interest_rate/4, period*4, price_after_five_years)/3
monthly_payment
```

# 3. Plot- apartment price/month; value of the investment

```python
freq = 12
period = 5
nper = period * freq
per = np.arange(1,nper+1,dtype=int)
nper1 = period *4
per1 = np.arange(1,nper1+1,dtype=int)
apartment_price = -npf.ppmt(percentage_per_year, per, nper, price_t0, -price_after_five_years)
plt.plot(apartment_price.cumsum(),label='Apartment price')
value_of_the_investment = -npf.ppmt(nominal_interest_rate/4,per,nper1*3, price_t0,-price_after_five_years)
plt.plot(value_of_the_investment.cumsum(),label='Value of the investment')
plt.legend()
plt.xlabel('time [months]')
plt.ylabel('Value [PLN]')
```
